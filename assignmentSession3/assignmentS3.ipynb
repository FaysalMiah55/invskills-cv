```python
# import libraties
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# load data
img_dir = '/content/drive/MyDrive/Colab Notebooks/invskills-cv/assignmentSession3/img_data'
images = os.listdir(img_dir)
images
```

```python
img_list = []
for img in os.listdir(img_dir):
  img_path = os.path.join(img_dir, img)
  img = cv2.imread(img_path)
  img_list.append(img)
```

```python
# display the images
fig, axes = plt.subplots(2, 5, figsize=(20, 5))
axes = axes.flatten()
for i, img in enumerate(img_list):
  axes[i].imshow(img)
  axes[i].axis('off')
plt.show()
```

Images looked blue because in openCV, the default color order is BGR

```python
# convert BGR to RGB color
fig, axes = plt.subplots(2, 5, figsize=(20, 5))
axes = axes.flatten()
for i, img in enumerate(img_list):
  axes[i].imshow(cv2.cvtColor(img_list[i], cv2.COLOR_BGR2RGB))
  axes[i].axis('off')
plt.show()
```

```python
# save images
save_images = '/content/drive/MyDrive/Colab Notebooks/invskills-cv/assignmentSession3'
os.makedirs(save_images, exist_ok=True)

for i, img in enumerate(img_list):
  cv2.imwrite(os.path.join(save_images, f'img_{i}.jpg'), img)
```
