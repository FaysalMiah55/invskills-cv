## Mandatory Task (MT):


#### Task-1: How image thresholding helps in each application mentioned in the materials?

Answer: Image thresholding convert grayscale images into binary images. This method apply in various applications.
- Object Detection
- Image Segmentation
- Optical Character Recognition
- Medical Imaging (like tumor detection, abnormal growths)
- Motion Detection

We will see the practical example furtur when completing detection task. Now we just see the example of thresholding.

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/invskills-cv/assignmentSession5/img1.webp', cv2.IMREAD_GRAYSCALE)
```

```python
# global thresholding
plt.imshow(img, cmap='gray')
```

```python
# ret, thresholded = cv.threshold(img, 230, 255, cv.THRESH_BINARY)

ret, thresholded = cv2.threshold(img, 230, 255, cv2.THRESH_BINARY)
plt.imshow(thresholded, cmap='gray')
print(ret)
```

```python
ret, thresholded = cv2.threshold(img, 230, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresholded, cmap='gray')
print(ret)
```

```python
ret, thresholded = cv2.threshold(img, 250, 255, cv2.THRESH_TRUNC)
plt.imshow(thresholded, cmap='gray')
print(ret)
```
